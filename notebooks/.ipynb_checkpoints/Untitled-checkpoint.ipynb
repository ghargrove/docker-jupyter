```python
from pathlib import Path
import numpy as np
import pandas as pd
import quandl as q
import matplotlib.pyplot as plt
from matplotlib import style

style.use('fivethirtyeight')

equities = ['MMM', 'FB']

q.ApiConfig.api_key = open('./quandl_key.txt').readline().rstrip()


## Write a function to calculate the sharpe ratio
def sharpe(series):
    '''Calculate the sharpe ratio for something funny'''
    start  = '2017-01-01'
    series = series[start:] 
    ## What will the params need to be? Dataframe?
    pct_change = series.pct_change()
    pct_change.fillna(0, inplace=True)
    
    print(series)
    print(pct_change)
    print('Average return: {}'.format(pct_change.mean()))
    print('Std Dev: {}'.format(pct_change.std()))
    
    return np.sqrt(250) * (pct_change.mean() / pct_change.std())



df = pd.DataFrame()

for e in equities:
    p_file = Path('./pickles/{}.pickle'.format(e))
    
    ## Load the equity data from the pickle
    if p_file.is_file():
        print('Loading {} data from pickle'.format(e))
        data = pd.read_pickle(str(p_file))
    else:
        print('Loading {} data from quandl API'.format(e))
        ticker = 'WIKI/' + e
        data = q.get(ticker)
        data.to_pickle(str(p_file))
        
    
    data = data[['Adj. Close']]
    data.rename(columns={'Adj. Close': e}, inplace=True)
    
    if df.empty:
        df = data
    else:
        df = df.join(data)

df.dropna(inplace=True)

for c in df.columns:
    df['{} % Change'.format(c)] = (df[c] - df[c][0]) / df[c][0] * 100
    
    print('{} Sharpe'.format(c))
    print(sharpe(df[c]))
    
# print(df.head())

df[['MMM % Change', 'FB % Change']].loc['2017-01-01':]

# plt.xlabel('Time')
# plt.ylabel('Return')
# plt.title('Returns of 2 stocks')
# plt.plot(df[['MMM % Change', 'FB % Change']].loc['2017-01-01':])
# plt.legend()
# plt.show()
```
